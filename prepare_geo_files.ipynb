```python
import json
from collections import defaultdict
```





Filerne er hentet via dataforsyningens API fx
`curl "https://api.dataforsyningen.dk/afstemningsomraader?format=geojson" -o afstemningsområder.geojson`

```python
afst = json.load(open("geo_files/raw/afstemningsområder.geojson", "r"))
```

Generer en fil med alle afstemningsområder per opstillingskreds

```python
groups = defaultdict(list)

# group features
for feature in afst["features"]:
    props = feature["properties"]
    key = f"{props['opstillingskredsnummer']}_{props['opstillingskredsnavn'].lower().replace(' ', '_')}"
    groups[key].append(feature)

# write one geojson per group
for key, features in groups.items():
    out = {
        "type": "FeatureCollection",
        "features": features
    }
    with open(f"geo_files/processed/opstillingskredse/{key}.geojson", "w", encoding="utf-8") as f:
         json.dump(out, f, ensure_ascii=False)
```
