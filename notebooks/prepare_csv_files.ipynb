```python
import pandas as pd
import os
```

```python
# Konverter Dansk tegnsætning til ASCII for at standardisere filnavne
def danish_to_ascii_filename(text):
    """Convert Danish characters to ASCII equivalents for safe filenames."""
    danish_to_ascii = {
        'æ': 'ae', 'ø': 'o', 'å': 'aa',
        'Æ': 'AE', 'Ø': 'O', 'Å': 'AA'
    }
    text = text.lower()
    for danish, ascii_equiv in danish_to_ascii.items():
        text = text.replace(danish, ascii_equiv)
    return text.replace(' ', '_')
```

```python
res = pd.read_csv("../data/struktureret/resultater_partier.csv")
```

```python
res
```

```python
# for each unique value in opstillingskreds_dagi_id, create an empty csv file with the columns kandidat,parti,valgt,stemmer and save it as opstillingskreds_dagi_id_opstillingskreds.csv in data/struktureret/opstillingskredse/personlige_stemmer
for opstillingskreds in res["opstillingskreds"].unique():
    opstillingskreds_id = res[res["opstillingskreds"] == opstillingskreds]["opstillingskreds_dagi_id"].iloc[0]
    df = pd.DataFrame(columns=["kandidat", "parti", "valgt", "stemmer"])
    df.to_csv(f"../data/struktureret/opstillingskredse/personlige_stemmer/{opstillingskreds_id}_{danish_to_ascii_filename(opstillingskreds)}.csv", index=False)
```

```python
# for each unique value in storkreds, create an empty csv file with the columns kandidat,parti,valgt,stemmer and save it as storkreds_id_storkreds.csv in data/struktureret/opstillingskredse/personlige_stemmer
for storkreds in res["storkreds"].unique():
    storkreds_id = res[res["storkreds"] == storkreds]["storkreds_nummer"].iloc[0]
    df = pd.DataFrame(columns=["kandidat", "parti", "valgt", "stemmer"])
    df.to_csv(f"../data/struktureret/storkredse/personlige_stemmer/{storkreds_id}_{danish_to_ascii_filename(storkreds)}.csv", index=False)
```

```python
# load in all the files in data/struktureret/opstillingskredse/kort, add a column called popup and turn it into ; separated csv files in data/struktureret/opstillingskredse/kort
for filename in os.listdir("../data/struktureret/storkredse/kort"):
    if filename.endswith(".csv"):
        df = pd.read_csv(f"../data/struktureret/storkredse/kort/{filename}")
        df["popup"] = ""
        df.to_csv(f"../data/struktureret/storkredse/kort/{filename}", index=False, sep=";")
```
