```python
# load in all the files in geo_files/processed/opstillingskredse and count the number of unique "valgsteder" in each file
antal_valgsteder = []
for file in os.listdir("../geo_files/processed/opstillingskredse"):
    with open(f"../geo_files/processed/opstillingskredse/{file}", "r") as f:
        opstillingskreds = file.split(".")[0]
        data = json.load(f)
        valgsteder = 0
        for feature in data["features"]:
            valgsteder += 1
        print(f"{opstillingskreds}: {valgsteder} valgsteder")
        antal_valgsteder.append({"opstillingskreds": opstillingskreds, "valgsteder": valgsteder})

df = pd.DataFrame(antal_valgsteder)

# sort by the opstillingskredsnavn after _
df["opstillingskredsnavn"] = df["opstillingskreds"].apply(lambda x: x.split("_")[1])
df_sorted = df.sort_values("opstillingskredsnavn").drop(columns=["opstillingskredsnavn"])


df_sorted.to_csv("../geo_files/processed/opstillingskredse_antal_valgsteder.csv", index=False)
```
