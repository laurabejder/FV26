```python
folder = "../data/resultater_2022/raw"
```

```python
# load in all the csv files in the folder and concatenate them into one dataframe

df = pd.DataFrame() # create an empty dataframe to hold the results

for file in os.listdir(folder):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(folder, file), sep=";")
        df['storkreds'] = file.split("_")[2] # extract storkreds from filename and add it as a column
        # remove .csv from storkreds
        df['storkreds'] = df['storkreds'].str.replace(".csv", "")
        if "resultater_2022" in locals():
            resultater_2022 = pd.concat([resultater_2022, df], ignore_index=True)
        else:
            resultater_2022 = df
```

```python
# standardiser partibogstaver, så A bliver til S, og så videre
bogstav_to_bogstav = {
    "A": "S",
    "B": "R",
    "C": "K",
    "D": "NB",
    "F": "SF",
    "H": "BP",
    "I": "LA",
    "M": "M",
    "O": "DF",
    "Q": "FG",
    "V": "V",
    "Æ": "DD",
    "Ø": "EL",
    "Å": "ALT",
    "(Uden for parti)": "UP",
}
```

```python
valgsted_resultater = (
    resultater_2022.groupby(["Afstemningsområde", "Partibogstav"])["Stemmetal"].sum().reset_index()
)

valgsted_resultater['procent_22'] = (round(
    valgsted_resultater['Stemmetal']
    / valgsted_resultater.groupby('Afstemningsområde')['Stemmetal'].transform('sum')
    * 100
,1))
```

```python
storkreds_resultater = (
    resultater_2022
    .groupby(['storkreds_dagi', 'storkreds', 'Partibogstav'])
    .agg({"Stemmetal": "sum"})
    .reset_index()
)

storkreds_resultater['procent_22'] = (round(
    storkreds_resultater['Stemmetal']
    / storkreds_resultater.groupby('storkreds_dagi')['Stemmetal'].transform('sum')
    * 100
,1))
```

```python
opstillingskreds_resultater = (
    resultater_2022
    .groupby(['opstillingskreds_dagi', 'Opstillingskreds', 'Partibogstav'])
    .agg({"Stemmetal": "sum"})
    .reset_index()
)

opstillingskreds_resultater['procent_22'] = (round(
    opstillingskreds_resultater['Stemmetal']
    / opstillingskreds_resultater.groupby('opstillingskreds_dagi')['Stemmetal'].transform('sum')
    * 100
,1))
```

```python
nationalt_resultater = (
    resultater_2022
    .groupby('Partibogstav')
    .agg({"Stemmetal": "sum"})
    .reset_index()
)

nationalt_resultater['procent_22'] = (round(
    nationalt_resultater['Stemmetal']
    / nationalt_resultater['Stemmetal'].sum()
    * 100
,1))
```

```python
# remove K from all files
valgsted_resultater = valgsted_resultater[valgsted_resultater["Partibogstav"] != "K"]
storkreds_resultater = storkreds_resultater[storkreds_resultater["Partibogstav"] != "K"]
opstillingskreds_resultater = opstillingskreds_resultater[opstillingskreds_resultater["Partibogstav"] != "K"]
nationalt_resultater = nationalt_resultater[nationalt_resultater["Partibogstav"] != "K"]
```

```python
# standardiser partibogstaver
for bogstav, standard in bogstav_to_bogstav.items():
    valgsted_resultater.loc[valgsted_resultater['Partibogstav'] == bogstav, 'Partibogstav'] = standard
    storkreds_resultater.loc[storkreds_resultater['Partibogstav'] == bogstav, 'Partibogstav'] = standard
    opstillingskreds_resultater.loc[opstillingskreds_resultater['Partibogstav'] == bogstav, 'Partibogstav'] = standard
    nationalt_resultater.loc[nationalt_resultater['Partibogstav'] == bogstav, 'Partibogstav'] = standard
```

```python
valgsted_resultater.to_csv("../data/resultater_2022/processed/afstemningssted_resultater.csv", index=False)
storkreds_resultater.to_csv("../data/resultater_2022/processed/storkreds_resultater.csv", index=False)
opstillingskreds_resultater.to_csv("../data/resultater_2022/processed/opstillingskreds_resultater.csv", index=False)
nationalt_resultater.to_csv("../data/resultater_2022/processed/nationalt_resultater.csv", index=False)
```
