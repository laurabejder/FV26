```python
import pandas as pd
```

```python
resultater_partier = pd.read_csv("data/struktureret/resultater_partier.csv")
res_opstillingskreds_2022 = pd.read_csv("data/resultater_2022/processed/opstillingskreds_resultater.csv")
```

```python
parti = "Æ"
parti1 = "DD"
```

```python
# filter resultater_partier to only include rows where parti is "O"
df = resultater_partier[resultater_partier["parti_bogstav"] == parti]
```

```python
df = df.groupby(['opstillingskreds_dagi_id',"opstillingskreds"]).agg({"stemmer": "sum",
                                                                "total_gyldige_stemmer":"sum"}).reset_index()
```

```python
df['procent_26'] = (round(df['stemmer'] / df['total_gyldige_stemmer'] * 100,1))
```

```python
df
```

```python
res_opstillingskreds_2022
res_opstillingskreds_2022 = res_opstillingskreds_2022[res_opstillingskreds_2022["Partibogstav"] == parti1]
```

```python
res_opstillingskreds_2022
```

```python
#join df and res_opstillingskreds_2022 on opstillingskreds_dagi_id and opstillingskreds_dagi
df_joined = pd.merge(df, res_opstillingskreds_2022, left_on='opstillingskreds_dagi_id', right_on='opstillingskreds_dagi', how='left')
```

```python
df_joined = df_joined[["opstillingskreds_dagi_id", "opstillingskreds", "procent_26", "procent_22"]]
```

```python
df_joined['diff_procent'] = round(df_joined['procent_26'] - df_joined['procent_22'],1)
```

```python
df_joined = df_joined.sort_values(by='diff_procent', ascending=False)
```
