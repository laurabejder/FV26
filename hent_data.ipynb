```python
import sys
import paramiko
import os
import re
import json
import pandas as pd

from config import HOST, PORT, USERNAME, PASSWORD, REMOTE_PATH
```

```python
transport = paramiko.Transport((HOST, PORT))
transport.connect(username=USERNAME, password=PASSWORD)
sftp = paramiko.SFTPClient.from_transport(transport)

# go into the data folder
sftp.chdir("/data/folketingsvalg-135-24-03-2026/valgdeltagelse")

# print all the folders in the data folder
files = sftp.listdir()
print("Folders in data folder:")
for file in files:
    # download the json files to data/raw/valgdeltagelse
    if file.endswith(".json"):
        print(file)
        sftp.get(file, os.path.join("data/raw/valgdeltagelse", file))
    
```

```python
# for each file in data/raw/valgresultater, load the json and print it
dts = []
for file in os.listdir("data/raw/valgresultater"):
    if file.endswith(".json"):
        with open(os.path.join("data/raw/valgresultater", file)) as f:
            data = f.read()
            data = json.loads(data)
            print(data.keys())
            dt = {
                "storkreds" : data["Storkreds"],
                "storkreds_nummer" : data["Storkredsnummer"],
                "opstillingskreds" : data["Opstillingskreds"],
                "opstillingskreds_dagi_id" : data["OpstillingskredsDagiId"],
                "afstemningsområde" : data["Afstemningsområde"],
                "afstemningsområde_dagi_id" : data["AfstemningsområdeDagiId"],
                "afstemningsområde_nummer": data["AfstemningsområdeNummer"],
                "gyldige_stemmer" : data["GyldigeStemmer"],
                "stemmeberettigede_vælgere" : data['AntalStemmeberettigedeVælgere']
            }
        dts.append(dt)
```

```python
# turn dts into a dataframe
df = pd.DataFrame(dts)

df['stemmeprocent'] = round(df['gyldige_stemmer'] / df['stemmeberettigede_vælgere'] *100,1)

# drop duplicates
df = df.drop_duplicates()
```
